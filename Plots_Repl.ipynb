```python
# Plot data, but only until time 100 and cut off header and index of dataframe
fig, ax = plt.subplots(3, 1, figsize=(9,9), sharex=True)
max = 100
time = np.arange(0, max)
ax[2].set_xlabel("time")
for i in range(3):
    ax[i].set_ylabel("opinion")
    # Exclude first column and row because they are the header and index of the dataframe
    for j in range(1,250):
        ax[i].plot(time, g[f"data{i+1}"][1:max+1,j], color='r', linewidth=0.2)
ax[0].set_title(f"$\epsilon$=0.2")
ax[1].set_title(f"$\epsilon$=0.6")
ax[2].set_title(f"$\epsilon$=1.0")
# Export figure
plt.savefig("D:\Daten mit Änderungen\Physik\Epidemics_Infodemics_Mobility\eps.png")
```
